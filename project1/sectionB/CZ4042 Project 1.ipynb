```python
## Modules used

import numpy as np
import pylab as plt
import tensorflow as tf
from sklearn.model_selection import KFold
```

## Question 1

```python
## Constants
SEED = 10
NUM_FEATURES = 8

# Model parameters
num_hidden_neuron = 30
batch_size = 32
decay = 0.001
learning_rate = 0.0000001

epochs = 501
```

```python
## Dataset Preparation

# Load and view the dataset
cal_housing = np.loadtxt('cal_housing.data', delimiter=',')
print("Dataset dimensions: " + str(cal_housing.shape)) # 20640 observations, 8 + 1 variables, last var is the dependent var
print("First row:")
print(cal_housing[0].view())

# Split into inputs and labels
X_data, Y_data = cal_housing[:,:8], cal_housing[:,-1] 
Y_data = (np.asmatrix(Y_data)).transpose()

# Shuffle data to break any symmetry
idx = np.arange(X_data.shape[0]) # no of rows
np.random.seed(SEED)
np.random.shuffle(idx)
print("\nShuffled incides: " + str(idx))
X_data, Y_data = X_data[idx], Y_data[idx] 

print("Check whether shuffling done correctly, next 2 rows should be the same:")
print(X_data[0].view())
print(cal_housing[20303].view())

# Read and divide data into test and train sets 
m = 3* X_data.shape[0] // 10
n = 7* X_data.shape[0] // 10
trainX, trainY = X_data[m:], Y_data[m:] # take the back 70% for trainset
testX, testY = X_data[:m], Y_data[:m] # take the front 30% for testset
print("\nSize of trainset: " + str(len(trainX)))
print("Size of testset: " + str(len(testX)))

# Normalize input
print("\nBefore normalization: " + str(trainX[0]))
trainX = (trainX- np.mean(trainX, axis=0))/ np.std(trainX, axis=0)
testX = (testX- np.mean(testX, axis=0))/ np.std(testX, axis=0)
print("Normalized: " + str(trainX[0]))
```

```python
## Model Creation

x = tf.placeholder(tf.float32, [None, NUM_FEATURES])
y_ = tf.placeholder(tf.float32, [None, 1])

weights = tf.Variable(tf.truncated_normal([NUM_FEATURES, num_hidden_neuron], stddev=1.0 / np.sqrt(NUM_FEATURES), dtype=tf.float32), name='weights')
biases = tf.Variable(tf.zeros([num_hidden_neuron]), dtype=tf.float32, name='biases')
y = tf.matmul(x, weights) + biases
y_relu = tf.nn.relu(y)

weights_out = tf.Variable(tf.truncated_normal([num_hidden_neuron, 1], stddev=1.0 / np.sqrt(num_hidden_neuron), dtype=tf.float32), name='weights')
biases_out = tf.Variable(tf.zeros([1]), dtype=tf.float32, name='biases')
y_out = tf.matmul(y_relu, weights_out) + biases_out

loss = tf.reduce_mean(tf.square(y_ - y_out))
cost = loss + decay * tf.nn.l2_loss(weights_out) + decay * tf.nn.l2_loss(weights)  # L2 regularization 

# Create the gradient descent optimizer with the given learning rate.
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
train_op = optimizer.minimize(cost)
error = tf.reduce_mean(tf.square(y_out - y_))
```

```python
## Session Creation

with tf.Session() as sess:

    sess.run(tf.global_variables_initializer())
    
    skf = KFold(n_splits=5, shuffle=True, random_state=10)
    
    val_err_list = []
    test_err_list = []
    test_predictions_list = []

    count = 0
    for train_index, val_index in skf.split(trainX, trainY):
        
        count += 1 
        print("Fold " + str(count))
        print("Training indexes: " + str(train_index))
        print("Validation indexes: " + str(val_index))
        
        val_err_list_fold = []

        trainX_train = trainX[train_index]
        trainY_train = trainY[train_index]

        trainX_val = trainX[val_index]
        trainY_val = trainY[val_index]

        for i in range(epochs):

            for start, end in zip(range(0, n, batch_size), range(batch_size, n, batch_size)):
                train_op.run(feed_dict={x: trainX_train[start:end], y_: trainY_train[start:end]})

            err = error.eval(feed_dict={x: trainX_val, y_: trainY_val})
            val_err_list_fold.append(err)

            if i % 100 == 0:
                print('iter %d: validation error %g'%(i, val_err_list_fold[i]))

        test_err = error.eval(feed_dict={x: testX, y_: testY})
        test_err_list.append(test_err)
        print('test error: %g'%(test_err))
        
        test_predictions = y_out.eval(feed_dict={x: testX, y_: testY})
        test_predictions_list.append(test_predictions)
        
        val_err_list.append(val_err_list_fold)
        print()
```

### 1(a) Plot validation errors against epoch

```python
average_val_error = []
err1, err2, err3, err4, err5 = zip(val_err_list)
for e1, e2, e3, e4, e5 in zip(err1[0], err2[0], err3[0], err4[0], err5[0]):
    average_err = (e1 + e2 + e3 + e4 + e5)/5
    average_val_error.append(average_err)
    

plt.figure(1)
plt.plot(range(epochs), average_val_error)
plt.title('Validation error against epoch number')
plt.xlabel(str(epochs) + ' iterations')
plt.ylabel('Validation Error')
plt.show()
```

```python
## Constants
SEED = 10
NUM_FEATURES = 8

# Model parameters
num_hidden_neuron_1 = 30
num_hidden_neuron_2 = 30
num_hidden_neuron_3 = 30
batch_size = 32
decay = 0.001
learning_rate = 0.000000001
keep_prob = 0.9

epochs = 501


## Model Creation (5 layers)

keep_prob = tf.placeholder(tf.float32)

x = tf.placeholder(tf.float32, [None, NUM_FEATURES])
y_ = tf.placeholder(tf.float32, [None, 1])

weights_1 = tf.Variable(tf.truncated_normal([NUM_FEATURES, num_hidden_neuron_1], stddev=1.0 / np.sqrt(NUM_FEATURES), dtype=tf.float32), name='weights')
biases_1 = tf.Variable(tf.zeros([num_hidden_neuron_1]), dtype=tf.float32, name='biases')
y_1 = tf.matmul(x, weights_1) + biases_1
y_relu_1 = tf.nn.relu(y_1)
y_dropout_1 = tf.nn.dropout(y_relu_1, keep_prob)

weights_2 = tf.Variable(tf.truncated_normal([num_hidden_neuron_1, num_hidden_neuron_2], stddev=1.0 / np.sqrt(num_hidden_neuron_1), dtype=tf.float32), name='weights')
biases_2 = tf.Variable(tf.zeros([num_hidden_neuron_2]), dtype=tf.float32, name='biases')
y_2 = tf.matmul(y_dropout_1, weights_2) + biases_2
y_relu_2 = tf.nn.relu(y_2)
y_dropout_2 = tf.nn.dropout(y_relu_2, keep_prob)

weights_3 = tf.Variable(tf.truncated_normal([num_hidden_neuron_2, num_hidden_neuron_3], stddev=1.0 / np.sqrt(num_hidden_neuron_2), dtype=tf.float32), name='weights')
biases_3 = tf.Variable(tf.zeros([num_hidden_neuron_3]), dtype=tf.float32, name='biases')
y_3 = tf.matmul(y_dropout_2, weights_3) + biases_3
y_relu_3 = tf.nn.relu(y_3)
y_dropout_3 = tf.nn.dropout(y_relu_3, keep_prob)

weights_out = tf.Variable(tf.truncated_normal([num_hidden_neuron_3, 1], stddev=1.0 / np.sqrt(num_hidden_neuron_3), dtype=tf.float32), name='weights')
biases_out = tf.Variable(tf.zeros([1]), dtype=tf.float32, name='biases')
y_out = tf.matmul(y_dropout_3, weights_out) + biases_out

loss = tf.reduce_mean(tf.square(y_ - y_out))
cost = loss + decay * tf.nn.l2_loss(weights_out) + decay * tf.nn.l2_loss(weights_3) + decay * tf.nn.l2_loss(weights_2) + decay * tf.nn.l2_loss(weights_1)  # L2 regularization 

# Create the gradient descent optimizer with the given learning rate.
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
train_op = optimizer.minimize(cost)
error = tf.reduce_mean(tf.square(y_out - y_))


## Session Creation

with tf.Session() as sess:

    sess.run(tf.global_variables_initializer())
        
    test_err_list = []

    trainX_train = trainX[train_index]
    trainY_train = trainY[train_index]

    trainX_val = trainX[val_index]
    trainY_val = trainY[val_index]

    for i in range(epochs):

        for start, end in zip(range(0, n, batch_size), range(batch_size, n, batch_size)):
            train_op.run(feed_dict={x: trainX_train[start:end], y_: trainY_train[start:end], keep_prob: 0.9})
        
        test_err = error.eval(feed_dict={x: testX, y_: testY, keep_prob: 1.0})
        test_err_list.append(test_err)

        if i % 100 == 0:
            print('iter %d: validation error %g'%(i, test_err))
            
plt.figure(1)
plt.plot(range(epochs), test_err_list)
plt.xlabel(str(epochs) + ' iterations')
plt.ylabel('Test Error')
plt.show()
```
